```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split, RandomizedSearchCV
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
import numpy as np

# Load train and test data
train_data = pd.read_csv(r'C:\Users\marco\OneDrive - usp.br\Área de Trabalho\mestrado\resultados\G_treino_imagens.csv', engine='python', sep=';')
test_data = pd.read_csv(r'C:\Users\marco\OneDrive - usp.br\Área de Trabalho\mestrado\resultados\G_teste_imagens.csv', engine='python', sep=';')

# Separate features and labels
X_train = train_data.drop('Grade', axis=1)
y_train = train_data['Grade']
X_test = test_data.drop('Grade', axis=1)
y_test = test_data['Grade']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
best_k = None
best_acc = 0.0
# Apply SMOTE to balance the training data
smote = SMOTE(k_neighbors=2)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)
best_k = None
best_acc = 0.0
# Apply SMOTE to balance the training data
smote = SMOTE(k_neighbors=2)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

for K in range(1,21):
    knn = KNeighborsClassifier(n_neighbors = K)
    cv_accuracy = cross_val_score(knn, X_train_resampled, y_train_resampled, cv=10, scoring='accuracy')
    mean_cv_accuracy = cv_accuracy.mean()
    dev_cv_accuracy = cv_accuracy.std()
    print(f'K = {K}: ({100*mean_cv_accuracy:.2f} ± {100*dev_cv_accuracy:.2f})%' )
    if best_acc < cv_accuracy.mean():
        best_acc = round(100*cv_accuracy.mean(),2)
        best_k = K
    
print(f'K = {best_k} -> acurácia = {best_acc}%')

best_acc = 0.0
best_n = None
for n in range(5,31,5):
    # Use SelectKBest to select the top k features
    k_best = SelectKBest(score_func=mutual_info_classif, k=n)  # Change k to the desired number of features
    X_train_k_best = k_best.fit_transform(X_train_resampled, y_train_resampled)
    X_test_k_best = k_best.transform(X_test)
    #Fit the Naive_Bayes 
    knn = KNeighborsClassifier(n_neighbors=best_k)
    fit = knn.fit(X_train_k_best, y_train_resampled)



    # Get the scores of the selected features
    feature_scores = k_best.scores_

    # Get the feature names
    feature_names = X_train.columns

    
    # Sort features by their scores
    sorted_indices = feature_scores.argsort()[::-1][:n]  # Select the top n features
    sorted_feature_names = [feature_names[i] for i in sorted_indices]
    sorted_feature_scores = feature_scores[sorted_indices]

    # Plot feature scores for the top n features (horizontal bars)
    plt.figure(figsize=(10, 6))
    plt.barh(range(n), sorted_feature_scores)
    plt.yticks(range(n), sorted_feature_names)
    plt.xlabel('Score')
    plt.ylabel('Features')
    plt.title(f'Top {n} Feature Scores')
    plt.gca().invert_yaxis()  # Invert the y-axis to display the highest score at the top
    plt.show()

    # Print the top k features
    selected_feature_indices = k_best.get_support(indices=True)
    selected_feature_names = [feature_names[i] for i in selected_feature_indices]
    print(f"Selected {len(selected_feature_names)} features: {', '.join(selected_feature_names)}")
    # Make predictions on the test data
    y_pred = fit.predict(X_test_k_best)

    cv_accuracy = cross_val_score(knn, X_train_k_best, y_train_resampled, cv=5, scoring='accuracy')
    mean_cv_accuracy = cv_accuracy.mean()
    dev_cv_accuracy = cv_accuracy.std()
    print(f'({100*mean_cv_accuracy:.2f} ± {100*dev_cv_accuracy:.2f})%' )

    # Accuracy with cross-validation
    cv_accuracy = cross_val_score(knn, X_test_k_best, y_test, cv=5, scoring='accuracy')
    mean_cv_accuracy = cv_accuracy.mean()
    dev_cv_accuracy = cv_accuracy.std()
    print(f'({100*mean_cv_accuracy:.2f} ± {100*dev_cv_accuracy:.2f})%' )
    if best_acc < cv_accuracy.mean():
        best_acc = cv_accuracy.mean()
        best_n = n
    
    # Classification report
    classification_rep = classification_report(y_test, y_pred)
    print(classification_rep)
    # Confusion matrix
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Normalize the confusion matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # Display the normalized confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_normalized, display_labels=[0, 1,2,3])
    disp.plot()
    plt.show()
    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
    import matplotlib.pyplot as plt

    # Calcular a matriz de confusão
    cm = confusion_matrix(y_test, y_pred, normalize='true')

    # Criar um heatmap
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, cmap='Blues', interpolation='nearest')
    plt.title('Matriz de Confusão Normalizada')
    plt.colorbar()

    # Adicionar rótulos aos eixos
    class_labels = ['0','1','2','3']  # Substitua pelos seus rótulos
    # Configure o tamanho da fonte
    a = True
    
    plt.rc('font', size=20)  # Defina o tamanho da fonte desejado
    
    tick_marks = np.arange(len(class_labels))
    plt.xticks(tick_marks, class_labels, rotation=45)
    plt.yticks(tick_marks, class_labels)

    # Adicionar os valores nas células
    for i in range(len(class_labels)):
        for j in range(len(class_labels)):
            plt.text(j, i, f"{cm[i, j]:.2f}", horizontalalignment="center", color="#ECAEAE")

    plt.tight_layout()
    plt.xlabel('Rótulos Preditos')
    plt.ylabel('Rótulos Verdadeiros')
    plt.show()
    plt.rc('font', size=10)
print(f'best n = {best_n}, best accurancy = {best_acc}')
```
